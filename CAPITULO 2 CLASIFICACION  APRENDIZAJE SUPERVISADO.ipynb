```python
def crear_scatter_iris(ejex,ejey):
    for t in range(3):
        if t == 0:
            c = 'r' # color = red
            marker = '>'
        elif t == 1:
            c = 'g' # color = green
            marker = 'o'
        elif t == 2: 
            c = 'b' # color = blue
            marker = 'x'
        plt.scatter(features[target == t,ejex],
                features[target == t,ejey],
                marker=marker,
                c=c)

    plt.xlabel(feature_names[ejex])
    plt.ylabel(feature_names[ejey])
```

```python
#print(target)
labels = target_names[target] #le asigna el name a cada elemento de target segun sea su valor 0 setosa, 1 versicolor etc
#print(labels)
petal_length = features[:,2] #armo un vector con los datos de la columna 2 de features que es length
#print(petal_length)
#print(labels == 'setosa')
is_setosa = (labels == 'setosa')# armo un vector en donde los q cumplen la condicion de setora son True y los demas False

#obtengo el maximo valor de longitud de esa variable en aquellos datos que son setosa, 
pl_setosa = petal_length[is_setosa] #es como un join entre los dos vectores y solo trae el petal length de los que son true,
                                     #es decir, de los que son setosa
print("Valores de petal_length para las setosas")
print(pl_setosa)
print("max petal_length", pl_setosa.max(), "cm para setosa")

#ahora traemos el minimo de los que no son setosa para determinar esa frontera de hasta donde puede ser setosa y cuando comienza
#otra especie
non_setosa = petal_length[~is_setosa]   # el simbolo ~ indica los que no son setosa, es decir hace join con los false
print("Valores de petal_length para las demas especies")
print(non_setosa)
print("min petal_length", non_setosa.min(), "cm para las demas especies")
```

### Lo anterior se hace para lo siguiente
    1. Tomamos el primer elemento de sepal length (cm) y lo convertimos en nuestro umbral. En este caso es 7 cm. 
    2. Luego, suponiendo que ese fuera el umbral, comparamos los demas valores de sepal length con ese umbral e indicamos cual es mayor que este.
    3. El resultado pred, es un vector que indica si el valor es mayor que el umbral elegido en ese momento
    
    Dado lo anterior, tenemos n vectores pred, uno por cada elemento que se eligio como umbral.
    
    Como esto se hace con cada columna, vamos a hacer esa comparacion por cada feature, buscando una lista de vectores a comparar con el vector de virginica. 
    
    El siguiente paso es comparar cada vector pred con el vector que contiene los valores de la especie virginica.


```python
best_acc = -1.0

sepal_le = np.array([],dtype=float)
sepal_wi = np.array([],dtype=float)
petal_le = np.array([],dtype=float)
petal_wi = np.array([],dtype=float)

for fi in range(features_new.shape[1]):
    print("Columna!!!!!!", fi, feature_names[fi])
    thresh = features_new[:,fi]
    for t in thresh:
        feature_i = features_new[:, fi] #por cada dato del vector hay un thresh
        #print("Comparando el vector", feature_i, " contra ", t) #Cada elemento se compara con los demas de su misma columna
        pred = (feature_i > t) # se construye un vector de T o F comparando cada valor con los demas de su misma columna
                               # si el elemento o valor es menor retorta true sino retorna false
        #print("Resutado: Dados los elementos:", feature_i, "los siguientes superan el umbral ", t)
        #print(pred)
        acc = (pred == is_virginica) #Si encontrara un vector acc que tuviera todos los elementos como true, encontrariamos 
                                     # un pred en donde el umbral elegido genera una clasificacion tal que las 
                                     # especies de flor serian como el target buscado, que en este caso es el vector is_virginica
        #print("resutado:")
        #if acc == True:
        #print(acc)
        #print("Con umbral" , t, "cm en", feature_names[fi])
        #print(acc.sum(axis=0)) #Imprimo cuantos elementos son iguales entre pred y is_virginica
        if fi == 0:
            sepal_le = np.append(sepal_le,acc.sum(axis=0)) #cuento los true
        elif fi == 1:
            sepal_wi = np.append(sepal_wi,acc.sum(axis=0))
        elif fi == 2:
            petal_le = np.append(petal_le, acc.sum(axis=0))
        else:
            petal_wi = np.append(petal_wi, acc.sum(axis=0))            
print(sepal_le.max())
print(sepal_wi.max())
print(petal_le.max())
indices, = np.where(petal_wi == petal_wi.max())
#print(indices)
#print(features_new[:,3]) #columa de petal width
print("Valores de petal width para los cuales el conteo de trues de la comparacion es", petal_wi.max(), "sobre un total de", features_new.shape[0], "elementos")
for i in indices:
    #print(i)
    print(features_new[:,3][i], "cm")
```

```python
best_acc = -1.0
for fi in range(features_new.shape[1]):
    thresh = features_new[:,fi]
    for t in thresh:
        feature_i = features_new[:, fi] #por cada dato del vector hay un thresh
        #print("Comparando el vector", feature_i, " contra ", t) #Cada elemento se compara con los demas de su misma columna
        pred = (feature_i > t) # se construye un vector de T o F comparando cada valor con los demas de su misma columna
                               # si el elemento o valor es menor retorta true sino retorna false
        #print("Resutado: Dados los elementos:", feature_i, "los siguientes superan el umbral ", t)
        #print(pred)
        acc = (pred == is_virginica).mean() # hace algo similar al conteo ya que saca el promedio, y el resultado es hacer la
                                            # la sumatoria de los valores, y como son true y false, sumara los valores 1
                                            # equivalente a lo que hice en np.append(sepal_le,acc.sum(axis=0), solo que saca el
                                            # promedio
        #print("Superando el umbral:", t, "cm para",feature_names[fi], "el promedio es", acc)
        rev_acc = (pred == ~is_virginica).mean() #ahora el promedio de los que no son virgnica 
        
        if rev_acc > acc: # Si el promedio de los elementos es mucho mas parecido en los que no son virginica entra
            reverse = True
            acc = rev_acc
        else:
            reverse = False
        
        if acc > rev_acc:
            best_acc = acc
            best_fi = fi
            best_t = t
            best_reverse = reverse
print(best_acc, best_fi, feature_names[best_fi], best_t, best_reverse)
```

```python
def is_virginica_test(fi, t, reverse, example):
    'Apply threshold model to a new example'
    #print(example[fi],"es mayor que", t)
    test = example[fi] > t
    #print(test)
    if reverse:
        test = not test
    return test
```

```python
example = features[~is_setosa] #los datos que no son setosa, es decir virginica y versicolor
virg_modelo = np.array([],dtype=bool)
for i in example:
    #print(i)
    resp = is_virginica_test(3, 1.7, False, i)
    #print(resp)
    virg_modelo = np.append(virg_modelo,resp)
print(virg_modelo)
print(is_virginica)
pres = (virg_modelo == is_virginica)
print(pres)
print(pres.sum(axis=0))
```
