```python
x = np.array([[1,2,3],[4,5,6]])
```

```python
print(x)
```

```python
from scipy import sparse
```

```python
eye = np.eye(4)
print(eye)
```

```python
#obtener las coordenadas de los elementos de la matrices anteriores, los valores ceros no se tienen en cuenta
m = sparse.csr_matrix(eye)
print(m)
m2 = sparse.csr_matrix(x)
print(m2)
```
